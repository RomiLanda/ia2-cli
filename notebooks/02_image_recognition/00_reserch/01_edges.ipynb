```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = "/resources/datasets/image_recognition/Screenshot from 2022-11-09 14-43-04.png"
image = cv2.imread(image_path)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blurred = cv2.GaussianBlur(gray, (3,3), 0)
```

```python
edged = cv2.Canny(blurred, 10, 100)
```

```python
plt.imshow(image)
```

```python
plt.imshow(edged)
```
